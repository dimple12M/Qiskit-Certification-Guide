# State Visualizations: To view the quantum state


A quantum state is 
1. state matrix œÅ(Hermitian matrix) or 
2. statevector |ùúì‚ü©(complex vector)


Different functions to visualize
- plot_state_city(quantum_state)
- plot_state_qsphere(quantum_state)
- plot_state_paulivec(quantum_state)
- plot_state_hinton(quantum_state)
- plot_bloch_multivector(quantum_state)

Reference:

https://qiskit.org/documentation/stubs/qiskit.visualization.plot_histogram.html#qiskit.visualization.plot_histogram

https://qiskit.org/documentation/tutorials/circuits/2_plotting_data_in_qiskit.html

---

```python
from qiskit import QuantumCircuit,BasicAer,execute
from qiskit.visualization import *
```

```python
qc=QuantumCircuit(2,2)
qc.h(0)
qc.x(1)
qc.draw(output="mpl")
```

```python
backend=BasicAer.get_backend('statevector_simulator')
job=execute(qc,backend)
result=job.result()
sv=result.get_statevector()
print(sv)
```

# plot_state_city: 

Reference:https://qiskit.org/documentation/stubs/qiskit.visualization.plot_state_city.html#qiskit.visualization.plot_state_city

- plots two 3d bar graphs (two dimensional) of the real and imaginary part of the density matrix rho


```python
plot_state_city(sv)
```

# plot_state_qsphere

Reference:

https://qiskit.org/documentation/stubs/qiskit.visualization.plot_state_qsphere.html#qiskit.visualization.plot_state_qsphere

- plots the qshpere representation of the quantum state
- shows the amplitude and phase of the statevector
- arrow represents the ball
- thickness of the arrow - amplitude
- color of the arrow- phase



```python
plot_state_qsphere(sv)
```

## Example to see the amplitude and phase

```python
qc=QuantumCircuit(2,2)
qc.x(0)
qc.h(0)
qc.h(1)
qc.s(0)

qc.draw(output="mpl")
```

```python
backend=BasicAer.get_backend('statevector_simulator')
job=execute(qc,backend)
result=job.result()
sv=result.get_statevector()
print(sv)
```

```python
plot_state_qsphere(sv)
```

# plot_state_paulivec

Reference:https://qiskit.org/documentation/stubs/qiskit.visualization.plot_state_paulivec.html#qiskit.visualization.plot_state_paulivec

# plot_state_hinton

# plot_bloch_multivector

Reference:https://qiskit.org/documentation/stubs/qiskit.visualization.plot_bloch_multivector.html#qiskit.visualization.plot_bloch_multivector

- plots the bloch sphere

```python
plot_bloch_multivector(sv)
```

Note:
- LHS shows the Hadamard gate
- RHS shows the X gate qubit

## Example: If the 2 qubits are entangled

```python
backend=BasicAer.get_backend('statevector_simulator')
job=execute(qc,backend)
result=job.result()
sv=result.get_statevector()
print(sv)
```

```python
plot_bloch_multivector(sv)
```
