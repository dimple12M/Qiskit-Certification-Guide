```python
from qiskit.visualization import *
from qiskit import QuantumCircuit,transpile
from qiskit import IBMQ
```

```python
provider=IBMQ.load_account()
provider
```

```python
provider.backends()
```

```python
backend=provider.get_backend("ibmq_lima")
backend
```

## <a id="GateMap">Gate map: </a>Plot the gate map of the device

https://qiskit.org/documentation/stubs/qiskit.visualization.plot_gate_map.html#qiskit.visualization.plot_gate_map

```python
plot_gate_map(backend)
```

## <a id="errorMap">Error map: </a>Plot the error map of the device

https://qiskit.org/documentation/stubs/qiskit.visualization.plot_error_map.html#qiskit.visualization.plot_error_map

```python
plot_error_map(backend)
```

## <a id="CircuitLayout">Circuit Layout : </a>Plot the layout of a transpiled  circuit  for the given backend

https://qiskit.org/documentation/stubs/qiskit.visualization.plot_circuit_layout.html#qiskit.visualization.plot_circuit_layout

```python
qc=QuantumCircuit(3)
qc.h(0)
qc.cx(0,1)
qc.z(1)
qc.draw(output="mpl")
```

```python
qc_transpile=transpile(qc,backend)
```

```python
qc_transpile.draw(output="mpl")
```

```python
plot_circuit_layout(qc_transpile,backend)
```
