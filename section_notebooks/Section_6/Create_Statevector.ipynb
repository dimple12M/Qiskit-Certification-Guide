```python
from qiskit import QuantumCircuit,Aer,transpile,execute
from qiskit.quantum_info import *
from qiskit.visualization import *
```

```python
qc=QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.draw(output="mpl")
```

# using Aer :
    1. statevector_simulator
    2. execute()


```python
backend=Aer.get_backend('statevector_simulator')

job=execute(qc,backend)

result=job.result()
sv=result.get_statevector()
```

```python
sv
```

```python
sv.data
```

```python
array_to_latex(sv.data)
```

```python
sv.draw("latex")
```

```python
sv.draw("qsphere")
```

# using Aer :
    1. statevector_simulator
    2. transpile and run

```python

qc_transpile=transpile(qc,backend)

job=backend.run(qc_transpile)
result=job.result()

sv=result.get_statevector()
```

```python
sv
```

```python
sv.data
```

```python
array_to_latex(sv.data)
```

```python
sv.draw("latex")
```

# using from_label and evolve from Statevector class

```python
sv_label=Statevector.from_label('++')
```

```python
sv_label.data
```

```python
qc=QuantumCircuit(2)
sv=Statevector.from_label('++')
sv_evolve=sv.evolve(qc)
sv_evolve
```

```python
sv_evolve.data
```

```python
sv
```

```python
sv_evolve.draw(output="latex")
```

```python
qc.draw(output="mpl")
```

```python
qc.initialize(sv_evolve)
```

# using the BasicAer: It has only run() 

```python
from qiskit import BasicAer
```

```python
qc=QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.draw(output="mpl")
```

```python
backend=BasicAer.get_backend('statevector_simulator')

qc_transpile=transpile(qc,backend)

job=backend.run(qc_transpile)
result=job.result()

sv=result.get_statevector()
```

```python
sv
```

```python
array_to_latex(sv.data)
```

# using the from_int from from Statevector class

- first parameter is the basis
- second parameter is the dimension

```python
# basis |0> and dim= 2^1=2
sv_label_0=Statevector.from_int(0,2**1)
sv_label_0.data
```

```python
sv_label_0.draw("latex")
```

```python
# basis 1(|01>) and dim= 2^2=4
sv_label=Statevector.from_int(1,2**2)
sv_label.data
```

```python
sv_label.draw("latex")
```

```python
# basis |0> and dim= 2^2=4
sv_label=Statevector.from_int(0,2**2)
sv_label.data
```

```python
sv_label.draw("latex")
```

```python
qc=QuantumCircuit(2)
qc.h(0)
qc.h(1)
sv_evolve=sv_label.evolve(qc)
sv_evolve
```

```python
sv_evolve.draw("latex")
```

```python
array_to_latex(sv_evolve.data)
```
