# Types of Bell States:

https://en.wikipedia.org/wiki/Bell_state

## 2-qubit bell state

4 types of maximally entagled states

INPUT 00:-----> $|\Phi^+\rangle = \frac{1}{\sqrt{2}}\left(|00\rangle + |11\rangle\right)$

INPUT 01:----->$|\Phi^-\rangle = \frac{1}{\sqrt{2}}\left(|00\rangle - |11\rangle\right)$

INPUT 10:----->$\vert\Psi^+\rangle = \frac{1}{\sqrt{2}}\left(\vert01\rangle + \vert10\rangle\right)$

INPUT 11:----->$\vert\Psi^-\rangle = \frac{1}{\sqrt{2}}\left(\vert01\rangle - \vert10\rangle\right)$.
 
 
  
## [3-qubit bell state](#3_qubitBell)
 

```python
from qiskit import QuantumCircuit,BasicAer,Aer,transpile,execute
from qiskit.visualization import *
```

# Example 1: 

input=00

--- 
### (|00> + |11>)/sqrt(2)

```python
qc=QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.measure_all()
qc.draw(output="mpl")
```

```python
backend=Aer.get_backend('qasm_simulator')
job=execute(qc,backend,shots=1024)
result=job.result()
counts=result.get_counts()
print(counts)
```

```python
# For the statevector ,no  measurements 
qc=QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.draw(output="mpl")
```

```python
backend=Aer.get_backend('statevector_simulator')
job=execute(qc,backend,shots=1024)
result=job.result()
sv=result.get_statevector()
sv.draw(output="latex")
```

```python
plot_state_qsphere(sv,show_state_phases=True)
```

```python
plot_state_city(sv)
```

# Additional Information


 
 $|+\rangle = \frac{1}{\sqrt{2}}\left(|0\rangle + |1\rangle\right)$
 
 
 $|-\rangle = \frac{1}{\sqrt{2}}\left(|0\rangle - |1\rangle\right)$
 
 $|\circlearrowleft\rangle = \frac{1}{\sqrt{2}}\left(|0\rangle - i|1\rangle\right)$

```python
qc=QuantumCircuit(1)
qc.h(0)
qc.draw(output="mpl")
```

```python
backend=Aer.get_backend('statevector_simulator')
job=execute(qc,backend,shots=1024)
result=job.result()
sv=result.get_statevector()
sv.draw(output="latex")
```
