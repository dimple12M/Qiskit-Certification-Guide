***
# Controlled Rotations
1. CRx
2. CRy
3. CRz
***

```python
from qiskit import *
from qiskit.quantum_info import Statevector
from math import pi
```

***
# Controlled RX gate:
***

https://qiskit.org/documentation/stubs/qiskit.circuit.library.CRXGate.html#qiskit.circuit.library.CRXGate

SYNTAX: crx(theta,control,target)

### CRX peforms rotation(X-axis) on the target qubit only when the control is in state|1>

```python
qc=QuantumCircuit(2)
qc.crx(pi/2,0,1)
qc.draw(output="mpl")
```

```python
backend = Aer.get_backend('unitary_simulator')

job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for CRx gate
print(result.get_unitary(qc, decimals=3))
```

```python
state = Statevector.from_instruction(qc)
state.draw(output='latex')
```


Reference: IBM Quantum Composer

```python
qc=QuantumCircuit(2)
qc.x(0)
qc.crx(pi/2,0,1)
qc.draw(output="mpl")
```

```python
state = Statevector.from_instruction(qc)
state.draw(output='latex')
```


Reference: IBM Quantum Composer
